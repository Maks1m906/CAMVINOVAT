## üè† –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ
1. –ù–∞–π–¥–∏—Ç–µ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ –≤–∞—à—É –ª—é–±–∏–º—É—é –∏–≥—Ä—É.  
2. –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –∏–≥—Ä—ã –≤–∞—à–µ–≥–æ –ª—é–±–∏–º–æ–≥–æ –∂–∞–Ω—Ä–∞.  
3. –ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ **3 –≤–æ–ø—Ä–æ—Å–∞** –æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–ª–µ–¥—É—é—â–µ–≥–æ —É—Ä–æ–∫–∞.

**–ò–¥–µ–∏ –≤–æ–ø—Ä–æ—Å–æ–≤:**
- –ö–∞–∫–∏–µ –∏–≥—Ä—ã –±—ã–ª–∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã –≤ –≥–æ–¥ –º–æ–µ–≥–æ —Ä–æ–∂–¥–µ–Ω–∏—è?
- –ö–∞–∫–∞—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ —Å–∞–º–∞—è —É—Å–ø–µ—à–Ω–∞—è –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º?
- –ö–∞–∫ –∏–∑–º–µ–Ω–∏–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º?


```python
import pandas as pd
import numpy as np

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã!")
print(f"–í–µ—Ä—Å–∏—è pandas: {pd.__version__}")
print(f"–í–µ—Ä—Å–∏—è numpy: {np.__version__}")

url = "https://gist.githubusercontent.com/designernatan/27da044c6dc823f7ac7fe3a01f4513ed/raw/d15b5c7d7a5efb38750b16ec935fc126ec9a6e79/vgsales.csv"

print("üîÑ –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ...")
df = pd.read_csv(url)

print("‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"–ó–∞–≥—Ä—É–∂–µ–Ω–æ –∏–≥—Ä: {len(df)}")

print("\n‚öΩ –≠–∫—à–µ–Ω –∏–≥—Ä—ã:")
action_games = df[df['Genre'] == 'Action']
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–∫—à–µ–Ω –∏–≥—Ä: {len(action_games)}")

top_sports = action_games.nlargest(3, 'Global_Sales')[['Name', 'Global_Sales', 'Platform', 'Publisher']]
print("–¢–æ–ø-3 —ç–∫—à–µ–Ω –∏–≥—Ä—ã:")
print(top_sports)
```


# üéÆ –î–µ–Ω—å 1 ‚Äî –ó–Ω–∞–∫–æ–º—Å—Ç–≤–æ —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –≤–∏–¥–µ–æ–∏–≥—Ä–∞—Ö
–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å! –í —ç—Ç–æ–º –Ω–æ—É—Ç–±—É–∫–µ –∑–∞–¥–∞–Ω–∏—è –∞–∫–∫—É—Ä–∞—Ç–Ω–æ **—Ä–∞–∑–¥–µ–ª–µ–Ω—ã –Ω–∞ —Ç–µ–∫—Å—Ç –∏ –∫–æ–¥**.

### ‚úÖ –ó–∞–¥–∞–Ω–∏–µ 1. –ü–µ—Ä–≤—ã–µ —à–∞–≥–∏
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ—Å—Ç—ã–µ –≤—ã–≤–æ–¥—ã –∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è. –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–æ–ø—É—Å–∫–∏.


```python
# –≠—Ç–æ –≤–∞—à –ø–µ—Ä–≤—ã–π –∫–æ–¥! –î–∞–≤–∞–π—Ç–µ –ø–æ–ø—Ä–∏–≤–µ—Ç—Å—Ç–≤—É–µ–º –º–∏—Ä –¥–∞–Ω–Ω—ã—Ö
print("–ü—Ä–∏–≤–µ—Ç, –º–∏—Ä –¥–∞–Ω–Ω—ã—Ö! üåç")

# –ó–ê–î–ê–ù–ò–ï: –î–æ–ø–∏—à–∏—Ç–µ –∫–æ–¥, —á—Ç–æ–±—ã –≤—ã–≤–µ—Å—Ç–∏ –≤–∞—à–µ –∏–º—è –∏ –≤–æ–∑—Ä–∞—Å—Ç
your_name = "Maksim"  # –í–ø–∏—à–∏—Ç–µ –≤–∞—à–µ –∏–º—è –≤ –∫–∞–≤—ã—á–∫–∞—Ö
your_age = 14     # –í–ø–∏—à–∏—Ç–µ –≤–∞—à –≤–æ–∑—Ä–∞—Å—Ç (—á–∏—Å–ª–æ –±–µ–∑ –∫–∞–≤—ã—á–µ–∫)

print(f"–ú–µ–Ω—è –∑–æ–≤—É—Ç {your_name}, –º–Ω–µ {your_age} –ª–µ—Ç")

# –ó–ê–î–ê–ù–ò–ï: –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –≤ –∫–∞–∫–æ–º –≥–æ–¥—É –≤—ã —Ä–æ–¥–∏–ª–∏—Å—å
current_year = 2025
birth_year = current_year - 14  # –í–º–µ—Å—Ç–æ ___ –≤—Å—Ç–∞–≤—å—Ç–µ –≤–∞—à –≤–æ–∑—Ä–∞—Å—Ç
print(f"–Ø —Ä–æ–¥–∏–ª—Å—è –≤ {birth_year} –≥–æ–¥—É")
```

### üß∞ –ó–∞–¥–∞–Ω–∏–µ 2. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
–ò–º–ø–æ—Ä—Ç–∏—Ä—É–π—Ç–µ `pandas` –∏ `numpy`. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≤–µ—Ä—Å–∏–∏.


```python
# –ó–ê–î–ê–ù–ò–ï: –ò–º–ø–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –Ω—É–∂–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
# –ü–æ–¥—Å–∫–∞–∑–∫–∞: –Ω–∞–º –Ω—É–∂–Ω—ã pandas –∏ numpy

import pandas as pd     # –í–º–µ—Å—Ç–æ ___ –Ω–∞–ø–∏—à–∏—Ç–µ pandas
import numpy as np     # –í–º–µ—Å—Ç–æ ___ –Ω–∞–ø–∏—à–∏—Ç–µ numpy

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã!")
print(f"–í–µ—Ä—Å–∏—è pandas: {pd.__version__}")
print(f"–í–µ—Ä—Å–∏—è numpy: {np.__version__}")
```


---
### ‚¨áÔ∏è –ó–∞–¥–∞–Ω–∏–µ 3. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
–ó–∞–≥—Ä—É–∑–∏—Ç–µ CSV –ø–æ URL –≤ DataFrame `df`.


```python
# URL –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –¥–∞–Ω–Ω—ã—Ö (—É–∂–µ –≥–æ—Ç–æ–≤ –¥–ª—è –≤–∞—Å)
url = "https://gist.githubusercontent.com/designernatan/27da044c6dc823f7ac7fe3a01f4513ed/raw/d15b5c7d7a5efb38750b16ec935fc126ec9a6e79/vgsales.csv"

# –ó–ê–î–ê–ù–ò–ï: –ó–∞–≥—Ä—É–∑–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –∏—Å–ø–æ–ª—å–∑—É—è pandas
print("üîÑ –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ...")
df = pd.read_csv(url)  # –í–º–µ—Å—Ç–æ ___ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ø—Ä–∞–≤–∏–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é

print("‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"–ó–∞–≥—Ä—É–∂–µ–Ω–æ –∏–≥—Ä: {len(df)}")
```


---
### üëÄ –ó–∞–¥–∞–Ω–∏–µ 4. –ü–µ—Ä–≤—ã–π –≤–∑–≥–ª—è–¥ –Ω–∞ –¥–∞–Ω–Ω—ã–µ
–ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞ –ø–µ—Ä–≤—ã–µ –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏.


```python
print("üîç –ü–µ—Ä–≤—ã–µ 5 –∏–≥—Ä:")
print(df.head())  # –ö–∞–∫—É—é —Ñ—É–Ω–∫—Ü–∏—é –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫?

print("\n" + "="*50)

print("üîç –ü–æ—Å–ª–µ–¥–Ω–∏–µ 5 –∏–≥—Ä:")
print(df.tail())  # –ö–∞–∫—É—é —Ñ—É–Ω–∫—Ü–∏—é –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ–∫?
```


---
### üß± –ó–∞–¥–∞–Ω–∏–µ 5. –°—Ç—Ä—É–∫—Ç—É—Ä–∞ DataFrame
–ü–æ–ª—É—á–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é, –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ —Ä–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞.


```python
print("üìã –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
df.info()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Å–µ—Ç–µ?

print("\n" + "="*30)

print("üìÇ –ù–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤:")
column_names = df.columns  # –ö–∞–∫ –ø–æ–ª—É—á–∏—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤?
print(column_names)

print("\n" + "="*30)

print("üìè –†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞:")
rows, columns = df.shape  # –ö–∞–∫–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–º–µ—Ä?
print(f"–°—Ç—Ä–æ–∫: {rows}")
print(f"–°—Ç–æ–ª–±—Ü–æ–≤: {columns}")
```


---
### üèÜ –ó–∞–¥–∞–Ω–∏–µ 6. –°–∞–º–∞—è –ø—Ä–æ–¥–∞–≤–∞–µ–º–∞—è –∏–≥—Ä–∞
–ù–∞–π–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º `Global_Sales` –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –∫–ª—é—á–µ–≤—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.


```python
print("üèÜ –ò—Å—Å–ª–µ–¥—É–µ–º —Ç–æ–ø –∏–≥—Ä—ã:")

max_sales_index = df['Global_Sales'].idxmax()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–∞—Ö–æ–¥–∏—Ç –∏–Ω–¥–µ–∫—Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è?
top_game = df.loc[max_sales_index]

print(f"–°–∞–º–∞—è –ø—Ä–æ–¥–∞–≤–∞–µ–º–∞—è –∏–≥—Ä–∞: {top_game['Name']}")
print(f"–ü—Ä–æ–¥–∞–∂–∏: {top_game['Global_Sales']} –º–ª–Ω –∫–æ–ø–∏–π")
print(f"–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞: {top_game['Platform']}")  # –ö–∞–∫–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ–¥–µ—Ä–∂–∏—Ç –ø–ª–∞—Ç—Ñ–æ—Ä–º—É?
print(f"–ñ–∞–Ω—Ä: {top_game['Genre']}")       # –ö–∞–∫–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ–¥–µ—Ä–∂–∏—Ç –∂–∞–Ω—Ä?
```


---
### üéÆ –ó–∞–¥–∞–Ω–∏–µ 7. –ü–ª–∞—Ç—Ñ–æ—Ä–º—ã –∏ –∂–∞–Ω—Ä—ã
–ü–æ–¥—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º, —Å–∞–º—É—é –ø–æ–ø—É–ª—è—Ä–Ω—É—é –ø–ª–∞—Ç—Ñ–æ—Ä–º—É –∏ —á–∏—Å–ª–æ –∂–∞–Ω—Ä–æ–≤.


```python
print("\nüéÆ –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã:")
unique_platforms = df['Platform'].nunique()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Å—á–∏—Ç–∞–µ—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è?
print(f"–í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º: {unique_platforms}")

platform_counts = df['Platform'].value_counts()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Å—á–∏—Ç–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è?
most_popular_platform = platform_counts.iloc[0]  # –ö–∞–∫ –ø–æ–ª—É—á–∏—Ç—å –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ Series?
print(f"–°–∞–º–∞—è –ø–æ–ø—É–ª—è—Ä–Ω–∞—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞: {most_popular_platform}")

print("\nüé≠ –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –∂–∞–Ω—Ä—ã:")
unique_genres = df['Genre'].nunique()  # –ö–∞–∫–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ–¥–µ—Ä–∂–∏—Ç –∂–∞–Ω—Ä—ã?
print(f"–í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤: {unique_genres}")
```


---
### üîé –ó–∞–¥–∞–Ω–∏–µ 8. –ü—Ä–æ—Å—Ç–∞—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—è
–û—Ç–±–µ—Ä–∏—Ç–µ –∏–≥—Ä—ã –Ω–∞ PS4 –∏ –∏–≥—Ä—ã –∂–∞–Ω—Ä–∞ *Sports*. –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø‚Äë3 –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –≤ —ç—Ç–æ–º –∂–∞–Ω—Ä–µ.


```python
print("\nüéÆ –ò–≥—Ä—ã –Ω–∞ PS4:")
ps4_games = df[df['Platform'] == 'PS4']  # –í–º–µ—Å—Ç–æ ___ –Ω–∞–ø–∏—à–∏—Ç–µ PS4
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –Ω–∞ PS4: {len(ps4_games)}")

print("\n‚öΩ –°–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∏–≥—Ä—ã:")
sports_games = df[df['Genre'] == 'Srorts']  # –ó–∞–ø–æ–ª–Ω–∏—Ç–µ: —Å—Ç–æ–ª–±–µ—Ü –∏ –∑–Ω–∞—á–µ–Ω–∏–µ
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã—Ö –∏–≥—Ä: {len(sports_games)}")

top_sports = sports_games.nlargest(3, 'Global_Sales')[['Name', 'Global_Sales']]  # –ö–∞–∫–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏?
print("–¢–æ–ø-3 —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∏–≥—Ä—ã:")
print(top_sports)
```

### ‚ûó –ó–∞–¥–∞–Ω–∏–µ 9. –ü—Ä–æ—Å—Ç–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—É–º–º—É, —Å—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω—É `Global_Sales`.


```python
print("\nüìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø—Ä–æ–¥–∞–∂:")
total_sales = df['Global_Sales'].sum()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Å—É–º–º–∏—Ä—É–µ—Ç –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è?
print(f"–û–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏ –≤—Å–µ—Ö –∏–≥—Ä: {total_sales:.1f} –º–ª–Ω –∫–æ–ø–∏–π")

average_sales = df['Global_Sales'].mean()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Å—á–∏—Ç–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ?
print(f"–°—Ä–µ–¥–Ω–∏–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä—ã: {average_sales:.2f} –º–ª–Ω –∫–æ–ø–∏–π")

median_sales = df['Global_Sales'].median()  # –ö–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Å—á–∏—Ç–∞–µ—Ç –º–µ–¥–∏–∞–Ω—É?
print(f"–ú–µ–¥–∏–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä—ã: {median_sales:.2f} –º–ª–Ω –∫–æ–ø–∏–π")
```

### üß™ –ó–∞–¥–∞–Ω–∏–µ 10. –¢–≤–æ—Ä—á–µ—Å–∫–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ

```python
print("\nüîç –í–†–ï–ú–Ø –î–õ–Ø –ò–°–°–õ–ï–î–û–í–ê–ù–ò–Ø!")

# –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏
min_sales_index = df['Global_Sales'].idxmin()  # –§—É–Ω–∫—Ü–∏—è, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∞—è –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
min_sales_game = df.loc[min_sales_index]
print(f"–ò–≥—Ä–∞ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏: {min_sales_game['Name']}")
print(f"–ü—Ä–æ–¥–∞–∂–∏: {min_sales_game['Global_Sales']} –º–ª–Ω –∫–æ–ø–∏–π")

# –°–∫–æ–ª—å–∫–æ –∏–≥—Ä –≤ 2008 –≥–æ–¥—É
games_2008 = df[df['Year'] == 2008]  # –ö–∞–∫–æ–π –≥–æ–¥ –Ω–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç?
print(f"–ò–≥—Ä –≤—ã–ø—É—â–µ–Ω–æ –≤ 2008 –≥–æ–¥—É: {len(games_2008)}")

# –í–∞—à–∞ –ª—é–±–∏–º–∞—è –∏–≥—Ä–∞
favorite_game_name = "Super Mario Bros."
favorite_game = df[df['Name'].str.contains(favorite_game_name, na=False)]

if not favorite_game.empty:
    print("\nüéÆ –ù–∞–π–¥–µ–Ω–∞ –≤–∞—à–∞ –ª—é–±–∏–º–∞—è –∏–≥—Ä–∞!")
    game_info = favorite_game.iloc[0]
    print(f"–ù–∞–∑–≤–∞–Ω–∏–µ: {game_info['Name']}")
    print(f"–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞: {game_info['Platform']}")
    print(f"–ñ–∞–Ω—Ä: {game_info['Genre']}")
    print(f"–ü—Ä–æ–¥–∞–∂–∏: {game_info['Global_Sales']} –º–ª–Ω –∫–æ–ø–∏–π")
else:
    print(f"\nüòî –ò–≥—Ä–∞ '{favorite_game_name}' –Ω–µ –Ω–∞–π–¥–µ–Ω–∞ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ")
    print("–ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–∏—Å–∫–∞—Ç—å –¥—Ä—É–≥—É—é –∏–≥—Ä—É!")
```
